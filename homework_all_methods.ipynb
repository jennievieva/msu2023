```python
import math
import numpy as np


def is_pos_def(x):
    return np.all(np.linalg.eigvals(x) > 0)


def gen_matrix(n):
    is_good_enough = False
    while not is_good_enough  :
        x = np.arange(1,n)
        V = []
        for i in range(n):
            y = np.random.randn(n)*2
            V.append(y)
        A = np.cov(V)
        is_good_enough = is_pos_def(A)
    return A
```

```python
n = int(input("Enter the size of the matrix: "))
```

## Метод Гаусса с numpy

```python
A = gen_matrix(n)
b = [1] * n
print(is_pos_def(A))
print(A)
```

```python
%%time
b = np.ones(n)
for k in range(0, n - 1):
    for i in range(k + 1, n):
        if A[i, k] != 0.0:
            lam = A[i, k] / A[k, k]
            A[i, k + 1:n] = A[i, k + 1:n] - lam * A[k, k + 1:n]
            b[i] = b[i] - lam*b[k]

x = b.copy()
for k in range(n - 1, -1, -1):
    x[k] = (b[k] - np.dot(A[k, k + 1:n], b[k + 1:n])) / A[k, k]

print(x)
```

## Метод Гаусса без numpy

```python
A = gen_matrix(n)
b = [1] * n
print(is_pos_def(A))
print(A)
```

```python
%%time
def gauss(A, b):
    for k in range(n):
        for i in range(k+1, n):
            s = A[i][k]/A[k][k]
            for j in range(n):
                   A[i][j] = A[i][j] - s * A[k][j]
    x = [0] * n
    for i in range(n-1, -1, -1):
        x[i] = b[i]
        for j in range(i+1, n):
            x[i] -= A[i][j] * x[j]
        x[i] /= A[i][i]

    return x


x = gauss(A, b)
print(x)
```

## Метод Холецкого c numpy

```python
A = gen_matrix(n)
b = [1] * n
print(is_pos_def(A))
print(A)
```

```python
%%time
def сholesky(A, b):
    L = np.linalg.cholesky(A)
    y = np.linalg.solve(L, b)
    x = np.linalg.solve(L.T, y)
    return x


x = сholesky(A, b)
print(x)
```

## Метод Холецкого без numpy

```python
A = gen_matrix(n)
b = [1] * n
print(is_pos_def(A))
print(A)
```

```python
%%time
def cholesky(A, b):
    n = len(A)
    L = [[0.0] * n for i in range(n)]
    for i in range(n):
        for k in range(i+1):
            tmp_sum = sum(L[i][j] * L[k][j] for j in range(k))
            if (i == k):
                L[i][k] = math.sqrt(A[i][i] - tmp_sum)
            else:
                L[i][k] = (1.0 / L[k][k] * (A[i][k] - tmp_sum))
    y = [0.0 for i in range(n)]
    for i in range(n):
        y[i] = b[i] / L[i][i]
        for k in range(i+1, n):
            y[i] -= L[k][i] / L[k][k] * y[k]
    x = [0.0 for i in range(n)]
    for i in range(n-1, -1, -1):
        x[i] = y[i]
        for k in range(i+1, n):
            x[i] -= L[k][i] * x[k]
        x[i] /= L[i][i]
    return x


x = cholesky(A, b)
print(x)
```
